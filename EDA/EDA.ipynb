```python
atrybuty_ilosciowe = df.select_dtypes(include=['number'])
atrybuty_ilosciowe.nunique()
```

```python
plt.figure(figsize=(15, 5))
plt.xticks(np.arange(0, 500001, step=50000))
sns.histplot(x=df["Cena"],bins=50, kde=True)
plt.show()
plt.figure(figsize=(15, 5))
sns.histplot(x=df["Cena"],bins=50, kde=True, log_scale=True)
plt.show()
```

Widać wyraźną lewoskośność histogramu, a zdecydowana większość aut kosztuje nie więcej niż 100000 złotych. Po przeskalowaniu na skalę logarytmiczną histogram staje się symetryczny.

```python
atrybuty_ilosciowe_bez_ceny=atrybuty_ilosciowe.drop(columns=["Cena"])
for i in atrybuty_ilosciowe_bez_ceny.columns:
    plt.figure(figsize=(15, 5))
    sns.histplot(atrybuty_ilosciowe[i], bins=35, kde=True)
    plt.show()
```

Histogramy dla parametrów wiek oraz przebieg są symetryczne, a dla mocy silnika — lewoskośny. Histogram pojemności silnika jest wielomodalny: najwyższy grzbiet widoczny jest w okolicach 2 litrów.

```python
for i in atrybuty_ilosciowe_bez_ceny.columns:
    plt.figure(figsize=(15, 5))
    sns.boxplot(x=atrybuty_ilosciowe_bez_ceny[i])
    plt.show()
for i in atrybuty_ilosciowe_bez_ceny.columns:
    plt.figure(figsize=(15, 5))
    sns.scatterplot(x=atrybuty_ilosciowe_bez_ceny[i], y=atrybuty_ilosciowe["Cena"])
    plt.show()
```

Większość analizowanych aut ma silniki o pojemności z zakresu 1,5-2 litrów oraz mocy od 105 do 165 KM, a także zostało wyprodukowane pomiędzy 11 a 19 lat temu. Przeciętne auto na OLXie ma przejechane ponad 200 tysięcy kilometrów.

Cena auta spada wraz ze wzrostem przebiegu oraz wieku, a rośnie razem z mocą i pojemnością silnika.

```python
plt.figure(figsize=(10, 10))
sns.heatmap(atrybuty_ilosciowe.corr(), annot=True, cmap="vlag")
```

Najbardziej skorelowane ze sobą są pary Moc-Pojemność, Moc-Cena oraz Cena-Wiek. Prawie nie ma korelacji pomiędzy mocą silnika a przebiegiem.

```python
atrybuty_kategoryczne = df.select_dtypes(object, bool)
atrybuty_kategoryczne.nunique()
```

```python
atrybuty_kategoryczne_bez_modeli = atrybuty_kategoryczne.drop(columns=["Model"])
for i in atrybuty_kategoryczne_bez_modeli.columns:
    category_order = atrybuty_kategoryczne_bez_modeli[i].value_counts().index.tolist()
    plt.figure(figsize=(7.5, 0.3 * atrybuty_kategoryczne_bez_modeli[i].nunique()))
    plot = sns.countplot(atrybuty_kategoryczne_bez_modeli[i], order=category_order)
    plot.bar_label(plot.containers[0])
    plt.show()
```

```python
for col in atrybuty_kategoryczne_bez_modeli.columns:
    category_order = atrybuty_kategoryczne_bez_modeli[col].value_counts().index
    values = df.groupby(col)["Cena"].agg(mean="mean", std="std", count="count").reindex(category_order)
    values["c_int_99"] = 2.576 * values["std"] / np.sqrt(values["count"])
    plt.figure(figsize=(7.5, 0.3 * len(values)))
    plot = sns.barplot(data=df, x="Cena", y=col, order=category_order, orient="h", errorbar=None)
    labels = values.apply(lambda x: f"{x['mean']:.0f} ± {x['c_int_99']:.0f}", axis=1)   
    plot.bar_label(plot.containers[0], labels=labels, color="black")
    plt.show()
```

```python
for col in atrybuty_kategoryczne_bez_modeli.columns:
    category_order = atrybuty_kategoryczne_bez_modeli[col].value_counts().index
    values = df.groupby(col)["Wiek"].agg(mean="mean", std="std", count="count").reindex(category_order)
    values["c_int_99"] = 2.576 * values["std"] / np.sqrt(values["count"])
    plt.figure(figsize=(7.5, 0.3 * len(values)))
    plot = sns.barplot(data=df, x="Wiek", y=col, order=category_order, orient="h", errorbar=None)
    labels = values.apply(lambda x: f"{x['mean']:.1f} ± {x['c_int_99']:.1f}", axis=1)   
    plot.bar_label(plot.containers[0], labels=labels, color="black")
    plt.show()
```

Wcześniej zauważona silna korelacja ceny z wiekiem znajduje swoje potwierdzenie w praktyce, tłumacząc, między innymi, dlaczego to białe auta są najdroższe.
